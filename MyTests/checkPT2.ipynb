```python
import numpy as np
import matplotlib.pylab as plt

#A = np.loadtxt('../full_hamiltonians/h6_1p00_chain.txt')
#A = np.loadtxt('ch4_sto6g_fullCI.txt')
#print(A.shape)
#print(A[0][0])
#E = -4.011042954447808739e+01
#
#firstC = np.loadtxt('firstC.txt')
#diag = np.loadtxt('diag.txt')

A = np.loadtxt('h2_6-31g_fullCI.txt')

E = A[0][0]
firstC = np.zeros(len(A[:,0]) - 1)
diag = np.zeros(len(A[:,0]) - 1)

for i in range(len(A[:,0]) - 1):
    firstC[i] = A[i+1][0]
    diag[i] = A[i+1][i+1]


ENPT2 = 0.

for i in range(len(diag)):
    print(f'The {i:6d}-th diagonal element is {diag[i]:.15f}')

print()
for i in range(len(diag)):
    denom = E - diag[i]
    de = firstC[i]*firstC[i]/denom
    print(f'The {i:6d}-th de is {de:.15f}')
    print(f'The {i:6d}-th sigma is {firstC[i]:.15f}')
    print(f'The {i:6d}-th denom is {denom:.15f}')
    print()
    ENPT2 = ENPT2 + de

print(ENPT2)
print(ENPT2 + E)


```
